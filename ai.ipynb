```python
import re
import nltk
import pdfplumber
from google.colab import files
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
from nltk.corpus import stopwords

```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z ]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z ]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
```

```python
def read_file(uploaded, filename):
    if filename.endswith(".txt"):
        return uploaded[filename].decode("utf-8", errors="ignore")

    elif filename.endswith(".pdf"):
        text = ""
        with pdfplumber.open(filename) as pdf:
            for page in pdf.pages:
                text += page.extract_text() + " "
        return text

    else:
        raise ValueError("Only .txt or .pdf files are supported")

```

```python
print("👋 Welcome to AI Resume Skill Gap Analyzer\n")

print("📄 Choose RESUME file (.txt or .pdf)")
resume_upload = files.upload()

print("\n📄 Choose JOB DESCRIPTION file (.txt or .pdf)")
job_upload = files.upload()

resume_file = list(resume_upload.keys())[0]
job_file = list(job_upload.keys())[0]

resume_text = read_file(resume_upload, resume_file)
job_text = read_file(job_upload, job_file)

print("\n✅ Files Loaded Successfully")
print("Resume:", resume_file)
print("Job Description:", job_file)
```

```python
clean_resume = clean_text(resume_text)
clean_job = clean_text(job_text)

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([clean_resume, clean_job])

similarity = cosine_similarity(vectors[0:1], vectors[1:2])[0][0]
matching_percentage = round(similarity * 100, 2)

resume_words = set(clean_resume.split())
job_words = set(clean_job.split())
missing_skills = job_words - resume_words

print("\n🎯 MATCHING RESULT")
print("Matching Percentage:", matching_percentage, "%")

print("\n❌ Missing Skills:")
for skill in missing_skills:
    print("-", skill)

print("\n💡 Improvement Suggestions:")
if matching_percentage < 40:
    print("• Resume is weakly matched. Add relevant skills.")
elif matching_percentage < 70:
    print("• Resume is moderately matched. Improve missing skills.")
else:
    print("• Resume is well matched. Minor improvements needed.")

print("• Skills to add:")
for skill in missing_skills:
    print("  -", skill)
```
